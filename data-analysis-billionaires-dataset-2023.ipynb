# Age

```python
# Fill the missing values with the mean of the "age" column
billionaires['age'].fillna(billionaires["age"].mean(), inplace = True)
```

```python
billionaires['age'].describe()
```

```python
plt.figure(figsize=(10, 6))  # Set the figure size

# Use Seaborn to create a histogram
sns.histplot(data=billionaires, x='age', kde=True, color='skyblue', bins=20)

# Add labels and title
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.title("Histogram of Age")

# Show the histogram
plt.show()
```

# Check the number of "Female" and "Male" billionaires

```python
gender_freq = billionaires['gender'].value_counts()
```

```python
gender_df = pd.DataFrame({"frequency": gender_freq, "percentage": gender_freq/(gender_freq.sum())*100})

gender_df
```

```python
labels = ['Male', 'Female']
colors = ['skyblue', 'lightcoral']
explode = (0, 0.1)  # To explode a slice (e.g., 'FM') for emphasis

# Create a pie chart
plt.figure(figsize=(8, 8))  # Set the figure size
plt.pie(gender_freq ,labels=labels, colors=colors, autopct='%1.1f%%', startangle=140, explode=explode)

plt.title("Distribution of Genders")
```

# The Self-Made vs Inherited Wealh.

### Checking the proportion of self-made and inherited wealth.

```python
seflmade_status = billionaires['selfMade'].value_counts()

seflmade_df = pd.DataFrame({"frequency": seflmade_status, "percentage": seflmade_status/(seflmade_status.sum())*100})

seflmade_df
```

### Data visualization for the self made status

```python
labels = ['Selfmade', 'Inherited']
colors = ['skyblue', 'lightcoral']
explode = (0, 0.1)  # To explode a slice (e.g., 'FM') for emphasis

# Create a pie chart
plt.figure(figsize=(8, 8))  # Set the figure size
plt.pie(seflmade_status ,labels=labels, colors=colors, autopct='%1.1f%%', startangle=140, explode=explode)

plt.title("Distribution of Selfmade status")
```

# Country

### Top 10 countries have the most billionaires:

```python
country_count = billionaires['country'].value_counts().head(10)

country_count
```

### Data Visualization for top 10 countries

```python
plt.figure(figsize=(10, 6))

country_count.plot(kind='bar', color = "skyblue")
plt.xlabel("Number of Billionaires")
plt.ylabel("Country")
plt.title("Top 10 Countries with the Most Billionaires")
plt.xticks(rotation=90)

plt.show()
```

# Category

# Top 10 cities with highest number of billionaires in the world

### Checking the proportion of self-made and inherited wealth.

```python
# Filter out rows with empty city values
filtered_billionaires = billionaires [ billionaires['city'] != ""]


# Group and count the occurrences of each city
city_count = filtered_billionaires['city'].value_counts().reset_index()
city_count.columns = ['city', 'count']

top_ten_city = city_count.head(10)
```

### Data visualization for the top cities

```python
plt.figure(figsize=(5, 3))

top_ten_city.plot(kind='bar', color = "yellow")
plt.xlabel("City count ")
plt.ylabel("City")
plt.title("Top 10 cities with the most billionaires")

plt.show()
```
